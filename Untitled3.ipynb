```python
# Merge trade data and order book data
# merged_data_BOO = pd.merge(data_BOO, order_BOO[['mid_0', 'mid_1']], how='outer', left_index=True, right_index=True)
merged_data_BOO = dd.merge(data_BOO, order_BOO[['mid_0', 'mid_1']], how='outer', left_index=True, right_index=True)
merged_data_BOO   # NaN checking asynchronous time series
```

```python
# Resapmle data with 1 second
data_BOO_resampled = data_BOO.resample('1S').last()
data_BOO_resampled['Time'] = data_BOO_resampled.index.strftime('%Y-%m-%d %H:%M:%S')
data_BOO_resampled['LogR'] = np.log(data_BOO_resampled['price'] / data_BOO_resampled['price'].shift(1))
data_BOO_resampled
```

```python
# Draw the plot for Log Returns
plt.figure(figsize=(20,10))
plt.plot(data_BOO_resampled['price'], label='BOO close price')
plt.plot(data_BOO_resampled['LogR'], label='BOO Log Return')
plt.title('Log Return for BOO')
plt.xlabel('Time')
plt.ylabel('Log Return')
plt.legend()
plt.show()
```

```python
# Resample data with 1 minute
data_BOO_resampled_min = data_BOO.resample('1T').last()
data_BOO_resampled_min['Time'] = data_BOO_resampled_min.index.strftime('%Y-%m-%d %H:%M:%S')
data_BOO_resampled_min['LogR'] = np.log(data_BOO_resampled_min['price'] / data_BOO_resampled_min['price'].shift(1))
data_BOO_resampled_min
plt.figure(figsize=(20, 10))
plt.plot(data_BOO_resampled_min['LogR'], label='BOO Log Return')
plt.title('Log Return for BOO')
plt.xlabel('Time')
plt.ylabel('Log Return')
plt.legend()
plt.show()
```

```python
# ACF PACF time stationary check
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plt.figure(figsize=(12, 3))

series_logR = data_BOO_resampled_min['LogR'].dropna()

plt.subplot(1,2,1)
plot_acf(series_logR, ax=plt.gca(), title='ACF for log return of BOO')
plt.subplot(1,2,2)
plot_pacf(series_logR, ax=plt.gca(), title='PACF for log return of BOO')
plt.show()
```

```python
# Volatility clustering
# compute realized volatility  1min for 30 minutes
import seaborn as sns

data_BOO_resampled_min['Realized_volatility'] = np.sqrt((data_BOO_resampled_min['LogR'] ** 2).rolling(window=30).sum()/30)
RV = data_BOO_resampled_min['Realized_volatility'].dropna()
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_BOO_resampled_min['Time'], y=RV, data= data_BOO_resampled_min)
plt.title('Scatter Plot of Realized Volatility with Clustering')
plt.xlabel('Squared Time Increments (\( \Delta t^2 \))')
plt.ylabel('Realized Volatility')
plt.show()
```
